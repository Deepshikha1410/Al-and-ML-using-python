#box plot

```python
fig, axes = plt.subplots(1,3 ,figsize=(15,4))

sns.boxplot(data = data_df,x='math score', ax= axes[0])
sns.boxplot(data = data_df,x='reading score',ax=axes[1])
sns.boxplot(data= data_df, x= 'writing score',ax=axes[2])
```

```python
Q1 = np.percentile(data_df['math score'],25)

Q3 = np.percentile(data_df['math score'],75)

outlier_dis =(Q3-Q1) * 1.5
outlier_dis
```

```python
outlier = data_df[data_df['math score'] < Q3 - outlier_dis].index
outlier
```

```python
from collections import Counter

outlier_list = []

outlier_list.extend( data_df[data_df['math score']< Q1-outlier_dis].index)

outlier_list.extend(data_df[data_df['writing score']< Q1 - outlier_dis].index)

outlier_list.extend(data_df[data_df['reading score']< Q1 - outlier_dis].index)

outlier_count = Counter(outlier_list)

outlier_count
```

```python
data_df['overall'] = (data_df['math score'] + data_df['reading score'] + data_df['writing score']) / 3.0 
```

```python
fig = plt.figure(figsize=(15, 12))

ax =  fig.add_subplot(221)

sns.boxplot(data = data_df, x='gender', y='overall',hue='test preparation course',ax=ax)

ax =fig.add_subplot(222)
sns.boxplot(data = data_df, x='gender', hue= 'parental level of education',y='overall',ax=ax)

ax=fig.add_subplot(223)
sns.boxplot(data = data_df,x='gender',hue='lunch',y='overall',ax=ax)

ax=fig.add_subplot(224)
sns.boxplot(data=data_df,x='gender',hue='race/ethnicity',y='overall',ax=ax)

plt.tight_layout()
```
